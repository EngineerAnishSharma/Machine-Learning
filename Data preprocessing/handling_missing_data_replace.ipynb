```python
import pandas as pd
import numpy as np
df = pd.read_csv("weather_data1.csv")
df
```

```python
new_df = df.replace(-99999,0)
new_df
```

```python
new_df = df.replace({
    'temperature':-99999,
    "event":"0",
    "windspeed":-99999
},np.NaN)
new_df
```

```python
new_df = df.replace([-99999,-8889],np.NAN)
new_df
```

```python
new_df = df.replace({
    -99999:np.NaN,
    'no event':np.NaN
})
new_df
```

```python
new_df = df.replace('[A-Za-z]','',regex=True)
new_df
```

```python
new_df = df.replace({
    'temperature':'[A-Za-z]',
    'windspeed':'[A-Za-z]'
},'',regex=True)
new_df
```
