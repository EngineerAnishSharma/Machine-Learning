```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("insurance_data.csv")
df
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.title("Insurance Data")
plt.xlabel("Age")
plt.ylabel("Is Insurance")
plt.scatter(df.age,df.bought_insurance,marker='+',color='red')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(df[['age']],df['bought_insurance'], test_size=0.1, random_state=10)
```

```python
from sklearn import linear_model
model = linear_model.LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.predict(X_test)
```

```python
import math
def sigmoid(x):
  return 1 / (1 + math.exp(-x))
```

```python
def prediction_function(age):
    z = 0.1274065 * age - -4.97339194 # 0.04150133 ~ 0.042 and -1.52726963 ~ -1.53
    y = sigmoid(z)
    return y
```
