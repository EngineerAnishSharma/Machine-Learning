Naive Bayes is a probabilistic classification algorithm based on Bayes' theorem. It's named "naive" because of the assumption of independence among the features, given the class. Despite the simplicity of this assumption, Naive Bayes classifiers often perform well in practice, particularly in tasks like text classification and spam filtering.

P(Y∣X)= 
P(X∣Y)⋅P(/P(X)
Use Gaussian Naive Bayes when dealing with continuous data and the assumption of a normal distribution is reasonable.

Use Multinomial Naive Bayes when dealing with discrete data, like word counts in text classificatio.

Use Bernoulli Naive Bayes when dealing with binary features or data represented as binary vects.Y)

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df
```

```python
inputs = df.drop(['PassengerId','Name','Parch','SibSp','Ticket','Cabin','Embarked'],axis='columns')
inputs.head()
```

```python
target = inputs.Survived
target
```

```python
dummies = pd.get_dummies(inputs.Sex)
dummies
```

```python
inputs = pd.concat([inputs,dummies], axis='columns')
inputs
```

```python
inputs.drop('Survived',axis='columns',inplace =True)
```

```python
inputs.head()
```

```python
inputs.columns[inputs.isna().any()]
```

```python
import math
inputs.Age=inputs['Age'].fillna(math.floor(inputs['Age'].mean()))
```

```python
inputs.head(10)
```

```python
inputs.columns[inputs.isna().any()]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(inputs.drop(['Sex','male'],axis='columns'),target,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.predict(X_test[:10])
```
