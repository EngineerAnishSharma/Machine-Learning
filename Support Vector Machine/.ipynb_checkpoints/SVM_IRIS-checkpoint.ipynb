```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
iris
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df
```

```python
df['target']= iris.target
```

```python
df[df.target==1].head()
```

```python
df[df.target==2].head()
```

```python
df[df.target==0].head()
```

```python
df['flower_name'] =df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
df0=df[df.target==0]
df1=df[df.target==1]
df2=df[df.target==2]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],color='g', marker="+")
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color='r', marker="*")
```

```python
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],color='g', marker="+")
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color='r', marker="*")
```

```python
X = df.drop(['target','flower_name'],axis='columns')
X.head()
```

```python
y= df.target
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.predict(X_test)
```
