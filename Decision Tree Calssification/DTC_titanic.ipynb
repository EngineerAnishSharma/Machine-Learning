```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("titanic.csv")
df
```

```python
new_df = df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
new_df
```

```python
from sklearn.preprocessing import LabelEncoder
le_sex = LabelEncoder()
```

```python
inputs = new_df.drop('Survived',axis='columns')
inputs
```

```python
inputs.Age=inputs['Age'].fillna(inputs.Age.mean())
inputs
```

```python
inputs['Sex'].replace({
    'male':1,
    'female':0
},inplace=True)
```

```python
inputs
```

```python
from sklearn.model_selection import train_test_split
```

```python
target  = df['Survived']
target
```

```python
X_train,X_test,y_train,y_test = train_test_split(inputs,target, test_size = 0.3,random_state=42)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.predict([[3,1,22,7]])
```
