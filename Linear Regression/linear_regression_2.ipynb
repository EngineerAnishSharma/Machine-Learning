```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df
```

```python
%matplotlib inline

plt.title("Canada per capita income")
plt.xlabel("Year")
plt.ylabel("Per Capita Income (US$)")  # Adjust the ylabel accordingly
plt.scatter(df['year'], df['per capita income (US$)'], color="g")
```

```python
year_col = df.drop('per capita income (US$)',axis = 'columns')
year_col

```

```python
reg = linear_model.LinearRegression()
reg.fit(year_col,df['per capita income (US$)'])
```

```python
reg.predict([[2020]])
```

```python
%matplotlib inline
plt.title("Canada per capita income")
plt.xlabel("area")
plt.ylabel("per capita income")
plt.plot(df['year'],reg.predict(df[['year']]), color = "r")
plt.scatter(df['year'],df['per capita income (US$)'])
```
